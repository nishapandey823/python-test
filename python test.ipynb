## Question 1.
As there there are five working days so windows size * 5

```python
#function for calculating rolling mean
def rolling_window(df,windowSize):
    return df.Close.rolling(window=windowSize*5).mean()
   
```

## Question 3  3.1 Volume shocks 

```python
#function for calculating volume shocks
def flag_df(df):

    if ((((df.unshifted-df.shifted)/df.shifted)*100)>10) :
        return True;      
    elif ((((df.unshifted-df.shifted)/df.shifted)*100)<-10):
        return False
    else :
        return np.nan
```

```python
#merging shifted and origional volumes for comparision.
df_merged_Infy=pd.DataFrame({"unshifted":df1.Volume,"shifted":df1.Volume.shift()})
df_merged_Infy['volumeShock'] = df_merged_Infy.apply(flag_df, axis = 1)
df_merged_Infy['volumeShock'].head()
```

## Q3.2 Price shocks

```python
df_merged_Infy['priceShock'] =(((df1.Close-df1.Close.shift())/df1.Close.shift())*100)>2
df_merged_Infy['priceShock'].head()
```

3.3 same as 3.2

## Q3.4  volumeless price movement

```python
df_merged_Infy['priceShockWithoutVolume'] = np.where((df_merged_Infy.priceShock) &   ( df_merged_Infy.volumeShock !=True), True, False)
df_merged_Infy['priceShockWithoutVolume'].head()
```

## Part 3 (data modelling - optional)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
prices = df1['Close'].tolist()
#dates = df1.index.tolist();

dates = np.array(df1.index)
unix_epoch = np.datetime64(0, 's')
one_second = np.timedelta64(1, 's')
features = (dates - unix_epoch) / one_second
features = np.reshape(features, (len(dates), 1))
prices = np.reshape(prices, (len(prices), 1))

X_train, X_test, y_train, y_test = train_test_split(features,prices,test_size=0.25)
linreg = LinearRegression()
linreg.fit(X_train,y_train)
Y_pred = linreg.predict(X_test)
Y_pred
```
